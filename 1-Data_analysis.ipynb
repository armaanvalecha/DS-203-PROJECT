```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import librosa
import glob
import os
```

```python
folder_path = 'MFCC-files/'
```

```python
results = []

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        # Read the CSV file
        df = pd.read_csv(file_path)
        # Calculate the number of columns divided by 86
        num_columns_divided = len(df.columns) / 86
        # Append the result to the list
        results.append((filename.split('-')[0], num_columns_divided))

# Create a DataFrame from the results
results_df = pd.DataFrame(results, columns=['Filename', 'Columns_Divided_By_86'])
```

```python
plt.figure(figsize=(14, 6))
plt.bar(results_df['Filename'], results_df['Columns_Divided_By_86'])
plt.xlabel('CSV Files')
plt.ylabel('Number of Columns Divided by 86')
plt.title('Number of Columns Divided by 86 for Each CSV File')
plt.xticks(rotation=0, fontsize=6)  # Adjust the fontsize here
plt.tight_layout()
plt.show()
```
